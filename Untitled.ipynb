## RDD Wordcount

```python
import findspark
findspark.init()
```

```python
#sparkcontext
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
```

```python
spark=SparkSession.builder.master("local[4]").appName("Wordcount_Rdd").getOrCreate()
```

```python
sc=spark.sparkContext
```

```python
sc.stop()
```

```python
veri_dosyasi= "C:\\Users\\ege_s\\Desktop\\pyspark\\omer_seyfettin_forsa_hikaye.txt"
```

```python
hikaye_rdd=sc.textFile(veri_dosyasi)
```

```python
hikaye_rdd.count() #kaç  satır oldugunu söyledi
```

```python
hikaye_rdd.take(5) #ilk beş taneisini okumak için
```

```python
#flatMap:herbir satırın herbir elemanının parçalanabilir hali üzerinde işlem yapıyor.
kelimeler= hikaye_rdd.flatMap(lambda satir: satir.split(" "))#herbir satır üzerine gir,boşluğa göre parçala
```

```python
kelimeler.take(10)
```

```python
kelime_sayilari= kelimeler.map(lambda kelime: (kelime,1)) #herbir kelimenin yanına 1 getir,tuple() olarak
```

```python
kelime_sayilari.take(5)
```

```python
kelime_sayilari_RBK=kelime_sayilari.reduceByKey(lambda x,y: x+y) 
#artık satırımızda 2 eleman oldugu için x,y dedik.key leri topladı.
#Yani o kelimeden kaç tane varsa yanına  onu yazdı.
```

```python
kelime_sayilari_RBK.take(10) #
```

```python
kelime_sayilari_RBK2=kelime_sayilari_RBK.map(lambda x:(x[1],x[0])) #rakamları başa,kelimeleri sona aldık
```

```python
kelime_sayilari_RBK2.take(5)
```

```python
kelime_sayilari_RBK2.sortByKey(False).take(20) #büyükten küçüğe dogru kelime sayılarını(tekrarlarını) yazdırdık.
```

## SparkContext,SparkConf,RDD Olusturma

```python
import findspark
findspark.init()  #pythonla sparkı bylusturuyoruz.
```

```python
#SparkContext:RDD ye giriş kapısı,1 den fazla yöntemle olusutrurlur.
#1.si SparkSession ile olusturma
```

```python
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf #ortam değişkenlerinin ayarlanamsına yardımcı olan bir nesne olusturuyor
```

```python
pyspark=SparkSession.builder \
.master("local[4]") \    #sparkın çalışma modeli. (*) koayrasak hepsini kullanabiliyoruz.burada local kullandık
.appName("RDD-Olusturmak") \ #sparkconf ile yapacagımız işleri elle yapıyoruz.executer:işi yapan işçi
.config("spark.executor.memory","4g") \  #kullanıcının kodu gönderdiği,sonuctan döndüğü pc
.config("spark.driver.memory","2g") \   #varsa getir yok ise yeniden yarat demek.
.getOrCreate()
```
